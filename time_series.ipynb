```python

def outlier(year):
  year_data=df[df['year'] == year]
  q1 = year_data['quantity'].quantile(0.25)
  q3= year_data['quantity'].quantile(0.75)
  IQR = q3 - q1

  lower_bound = q1 - 1.5 * IQR
  upper_bound = q3 + 1.5 * IQR

  # Identify outliers
  outliers_first_year = year_data[(year_data['quantity'] < lower_bound) | (year_data['quantity'] > upper_bound)]
  outliers_first_year


  s=outliers_first_year.index

  for i in s:
    interpolated_value = (df.at[i-1, 'quantity'] + df.at[i+1, 'quantity']) / 2

# Replace outlier value with interpolated value
    df.at[i, 'quantity'] = interpolated_value
```
